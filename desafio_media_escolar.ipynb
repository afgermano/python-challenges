## Desafio da Alura

**Percorrer a lista usando um loop while e adicionar 1 nota a cada valor, exceto quando a soma ultrapassar 10, caso em que o valor máximo deve ser 10.**

---

### Descrição da Funcionalidade

O sistema permite que o usuário escolha se deseja adicionar uma **nota extra** às médias dos alunos, informando também o **valor dessa nota adicional**, que será somado à nota atual de cada aluno.

### Regras Implementadas

- O usuário pode responder **"Sim" ou "Não"** para a adição da nota extra.  
  O código trata diferentes variações de escrita (ex: `sim`, `SIM`, `Sim`, `sIm`), padronizando a resposta para `"sim"` antes da verificação no `if`.

- O sistema valida a **nota máxima permitida (10)**.  
  Caso a soma da nota atual com a nota extra ultrapasse esse valor, a nota final do aluno será limitada a **10**.

- Ao final da execução, o programa exibe o **nome de cada aluno** juntamente com sua **nota final**.

```python
lista_nomes = ["Ana Souza", "Livia Carvalho", "Paola Oliveira", "Italo Oliveira", "Joao Ferreira"]
lista_notas = [3.5, 4.7, 7.5, 10.0, 8.8]

resposta = input("Deseja inserir a nota extra do trabalho? (Sim/Não): ").strip().lower()

if resposta == "sim":
  nota = float(input("Digite o valor da nota extra: "))
  for i in range(len(lista_notas)):
        if lista_notas[i] <= 9:
           lista_notas[i] += nota

        if lista_notas[i] > 10:
           lista_notas[i] = 10

i = 0

while i < len(lista_nomes):
    print(f"{lista_nomes[i]} - Nota final: {lista_notas[i]}")
    i += 1
```
